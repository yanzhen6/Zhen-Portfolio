## Data Cleaning and Wrangling

```python
import re
import string
```

```python
def clean_text(text):
    #Make text lowercase, remove punctuations, and remove numbers
    text = text.lower()
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub('[0-9]+','', text)
    return text
```

```python
dat1['tweetTextOriginal'] = dat1['tweetTextOriginal'].map(clean_text)
dat1
```

## Organizing Data

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(stop_words='english')

tweetCV = cv.fit_transform(dat1['tweetTextOriginal'])
tweetCV_dtm = pd.DataFrame(tweetCV.toarray(), columns = cv.get_feature_names())

tweetCV_dtm.index = dat1['tweetTextOriginal'].index
tweetCV_dtm
```

## Topic Modeling

```python
from gensim import matutils, models
import scipy.sparse
```

```python
tdm = tweetCV_dtm.transpose()
sparse_counts = scipy.sparse.csr_matrix(tdm)
corpus = matutils.Sparse2Corpus(sparse_counts)
id2word = dict((v, p) for p, v in cv.vocabulary_.items())
```

```python
lda = models.LdaModel(corpus=corpus, id2word=id2word, num_topics=3, passes=20)
lda.print_topics()
```

```python
lda = models.LdaModel(corpus=corpus, id2word=id2word, num_topics=4, passes=20)
lda.print_topics()
```

```python
lda = models.LdaModel(corpus=corpus, id2word=id2word, num_topics=5, passes=10)
lda.print_topics()
```
